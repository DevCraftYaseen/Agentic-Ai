### Langgraph Sequential Graph - BMI Calculator

```python
from langgraph.graph import StateGraph, START , END
from typing import TypedDict
```

```python
# Define Initial State
class BMIState(TypedDict) :
    weight_kg : float
    height_m : float
    bmi : float
    bmi_category : str
```

```python
# Define Functions
# 1 . Calculate BMI
def calculate_bmi_score(state : BMIState) -> BMIState:
    weight = state['weight_kg']
    height = state['height_m']
    
    bmi_score = weight / (height ** 2)
    
    state['bmi'] = round(bmi_score)
    
    return state

# 2. Categorize BMI
def categorize_bmi(state : BMIState) -> BMIState:
    bmi_score = state['bmi']
    if bmi_score <= 18.4:
        state['bmi_category'] = 'Underweight'
    elif bmi_score <= 24.9:
        state['bmi_category'] = 'Normal weight'
    elif bmi_score <= 29.9:
        state['bmi_category'] = 'Overweight'
    else:
        state['bmi_category'] = 'Obesity'
    
    return state
```

```python
# Define The Graph
graph = StateGraph(BMIState)

# Add Nodes
graph.add_node('calculate_bmi', calculate_bmi_score)
graph.add_node('categorize_bmi', categorize_bmi)

#Add Edges
graph.add_edge(START, 'calculate_bmi')
graph.add_edge('calculate_bmi', "categorize_bmi")
graph.add_edge("categorize_bmi", END)

# Compile The Graph
compiled_workflow = graph.compile()
```

```python
# Execute The Graph
initial_state = {"weight_kg" : 82 , "height_m" : 1.7526}
final_state = compiled_workflow.invoke(initial_state)
print(final_state)
```
